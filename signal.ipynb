```python
import numpy as np
import matplotlib.pyplot as plt 
import math

def signal(t,T,a):
             return math.cos(2*math.pi*a*(1-t**2)/T)
            # return 1 + math.cos(2*math.pi*t/T) + 0.5*math.cos(4*math.pi*t/T) + math.sin(6*math.pi*t/T) + 0.2*math.cos(20*math.pi*t/T)
            # return math.sin(2*math.pi*t/T)

```

```python
import matplotlib.pyplot as plt

a = 1000
T = 0.1
N = 1000

x = np.linspace(0,T,N, endpoint=True)
y = np.zeros(N, dtype=float)

fe1 = 100.0
te1 = 1.0/fe1
N1 = int(T*fe1)
t1 = np.arange(0,N1)*te1


for i in range(0,len(x)-1):
    y[i]=signal(x[i],T,a)

#plt.plot(x, z)
plt.plot(x, y)
plt.show()
```

```python
A = np.fft.fft(y)
plt.plot(x,A)
plt.show()
```

```python
freq=np.fft.fftfreq(N,0.1)
plt.plot(freq, A.real, label= "Rel")
plt.plot(freq, A.imag, label ="Image")
```

```python
signal_reconstruit = np.fft.ifft(A)
plt.plot(x,signal_reconstruit)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres du signal
f = 10  # Frquence du signal en Hz
A = 1   # Amplitude du signal

# Paramtres de l'chantillonnage
fe = 20  # Frquence d'chantillonnage en Hz
T = 1/fe # Dure d'un chantillon en s

# Gnration du signal sinusodal
t = np.arange(0, 1, T) # Echantillonnage du signal
x = A * np.cos(2 * np.pi * f * t)

# Affichage du signal
plt.plot(t, x)
plt.title('Signal sinusodal de frquence f = 10 Hz')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude')
plt.show()




```

```python
# Paramtres du signal
f = 10  # Frquence du signal en Hz
A = 1   # Amplitude du signal

# Paramtres de l'chantillonnage
fe = 20  # Frquence d'chantillonnage en Hz
T = 1/fe # Dure d'un chantillon en s

# Gnration du signal sinusodal
t = np.arange(0, 1, T) # Echantillonnage du signal
x = A * np.cos(2 * np.pi * f * t)

# Affichage du signal chantillonn
plt.stem(t, x, 'r', markerfmt='ro', label='Signal chantillonn')

# Gnration du signal aliasing
fa = f - fe # Frquence du signal aliasing en Hz
t_aliasing = np.arange(0, 1, T) # Echantillonnage du signal aliasing
x_aliasing = A * np.cos(2 * np.pi * fa * t_aliasing)

# Affichage du signal aliasing
plt.plot(t_aliasing, x_aliasing, 'b', label='Signal aliasing')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gnration du signal
n = np.arange(0, 64)
x = np.cos(np.pi * n / 8)

# Calcul de la DFT du signal
X = np.fft.fft(x)

# Affichage du signal et de sa DFT
plt.subplot(2, 1, 1)
plt.plot(n, x)
plt.xlabel('n')
plt.ylabel('x[n]')

plt.subplot(2, 1, 2)
plt.plot(n, np.abs(X))
plt.xlabel('n')
plt.ylabel('|X(k)|')

plt.show()
```

```python
frequencies = np.fft.fftfreq(len(x))

# Affichage de la DFT dans le domaine des frquences
plt.plot(frequencies, np.abs(X))
plt.xlabel('Frquence (Hz)')
plt.ylabel('Amplitude')
plt.show()
```
